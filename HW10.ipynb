```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = df.drop(columns = ['PassengerId', 'Pclass', 'Name', 'SibSp', 'Cabin', 'Fare', 'Ticket'])
```

```python
df
```

```python
mapping = {'male':0, 'female':1}
df['Sex'] = df['Sex'].map(mapping)
```

```python
df
```

```python
set(df['Embarked'])
```

```python
df = df.dropna()
```

```python
set(df['Embarked'])
```

```python
mapping = {'C':0, 'Q':1, 'S':2}
df['Embarked'] = df['Embarked'].map(mapping)
```

```python
df.head()
```

```python
x = df.iloc[:, 1:]
y = df.iloc[:, 1]
```

```python
x
```

```python
y
```

```python
x = x.values
y = y.values
```

```python
x
```

```python
scaler = StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 15, random_state = 2025)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
round(np.sum(knn.predict(X_test)==y_test)/len(y_test), 2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
round(np.sum(rf.predict(X_test)==y_test)/len(y_test), 2)
```

```python
knn_1 = KNeighborsClassifier(1)
knn_3 = KNeighborsClassifier(3)
knn_5 = KNeighborsClassifier(5)
knn_10 = KNeighborsClassifier(10)
knn_20 = KNeighborsClassifier(20)
rf_1 = RandomForestClassifier(1)
rf_3 = RandomForestClassifier(3)
rf_5 = RandomForestClassifier(5)
rf_10 = RandomForestClassifier(10)
rf_20 = RandomForestClassifier(20)
knn_1.fit(X_train, y_train)
knn_3.fit(X_train, y_train)
knn_5.fit(X_train, y_train)
knn_10.fit(X_train, y_train)
knn_20.fit(X_train, y_train)
rf_1.fit(X_train, y_train)
rf_3.fit(X_train, y_train)
rf_5.fit(X_train, y_train)
rf_10.fit(X_train, y_train)
rf_20.fit(X_train, y_train)
```

```python
print(f'knn_1: {round(np.sum(knn_1.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'knn_3: {round(np.sum(knn_3.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'knn_5: {round(np.sum(knn_5.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'knn_10: {round(np.sum(knn_10.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'knn_20: {round(np.sum(knn_20.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'rf_1: {round(np.sum(rf_1.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'rf_3: {round(np.sum(rf_3.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'rf_5: {round(np.sum(rf_5.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'rf_10: {round(np.sum(rf_10.predict(X_test)==y_test)/len(y_test), 2)}')
print(f'rf_20: {round(np.sum(rf_20.predict(X_test)==y_test)/len(y_test), 2)}')
```
