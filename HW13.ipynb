```python
import numpy as np
```

```python
y_observed = [0.06, 0.8, 0.11, 0.65, 0.34, 0.5, 0.74, 0.62, 0.4]
y_pred = [0.05, 0.9, 0.1, 0.6, 0.2, 0.8, 0.3, 0.7, 0.5]
```

```python
def mse(y_observed, y_pred):
    y_observed = np.array(y_observed)
    y_pred = np.array(y_pred)
    sum = 0
    for i in range(y_observed.size):
        sum+=np.power((y_observed[i]-y_pred[i]), 2)
    return (sum/y_observed.size);
    
```

```python
print(mse(y_observed, y_pred))
```

```python
def mae(y_observed, y_pred):
    y_observed = np.array(y_observed);
    y_pred = np.array(y_pred);
    sum = 0;
    for i in range(y_observed.size):
        sum+=np.absolute(y_observed[i]-y_pred[i]);
    return sum/y_observed.size;
```

```python
print(mae(y_observed, y_pred))
```

```python
def dl(n):
    model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
    model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
    xs  = np.array([0,1,2,3,4,5,6], dtype = float)
    ys = np.array([50,100,150,200,250,300,350], dtype = float)
    model.fit(xs, ys, epochs = 500)
    return model.predict(np.array(n))
```

```python
def house_model_scaled_MSE(y_new):
    xs = np.array([0, 1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([0.5, 1, 1.5, 2, 2.5, 3, 3.5], dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='mean_squared_error')
    
    model.fit(xs, ys, epochs=500)
    
    return model.predict(y_new)[0][0]*100
```

```python
def house_model_scaled_MAE(y_new):
    xs = np.array([0, 1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([0.5, 1, 1.5, 2, 2.5, 3, 3.5], dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='mae')
    
    model.fit(xs, ys, epochs=500)
    
    return model.predict(y_new)[0][0]*100
```

```python
def house_model_scaled_MSLE(y_new):
    xs = np.array([0, 1, 2, 3, 4, 5, 6], dtype=float)
    ys = np.array([0.5, 1, 1.5, 2, 2.5, 3, 3.5], dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='msle')
    
    model.fit(xs, ys, epochs=500)
    
    return model.predict(y_new)[0][0]*100
```

```python
import pandas as pd
data = pd.read_csv('housing.csv')
data
```

```python
data = data.dropna()
```

```python
data
```

```python
def housing(y_new):
    xs = np.array(data.area, dtype=float)
    ys = np.array(data.price, dtype=float)
    
    model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    
    model.compile(optimizer='sgd', loss='mae')
    
    model.fit(xs, ys, epochs=500)
    
    return model.predict(y_new)[0][0]*100
```

```python
housing(np.array([[5000]]))
```
