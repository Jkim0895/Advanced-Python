```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['class'] = wine.target
```

```python
df.head()
```

```python
df_notarget = df.iloc[:, :-1]
```

```python
df_notarget
```

```python
plt.figure(figsize = (7,7))
sns.clustermap(df_notarget.corr(method = 'pearson'), cmap = 'viridis', annot = True)
plt.title('Correlation Coefficient between wine features')
plt.show()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
x =  x.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 2025)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
svm.predict(X_test) == y_test
```

```python
np.mean(svm.predict(X_test) == y_test)
```

```python
svm_multihyp1 = SVC(kernel = 'poly', gamma = 10, C = 10)
svm_multihyp2 = SVC(kernel = 'linear', gamma = 10, C = 10)
svm_multihyp3 = SVC(kernel = 'rbf', gamma = 10, C = 10)
svm_multihyp1.fit(X_train, y_train)
svm_multihyp2.fit(X_train, y_train)
svm_multihyp3.fit(X_train, y_train)
print(f'prediction accurracy of polykernel, gamma10, C10 is {np.mean(svm_multihyp1.predict(X_test) == y_test)}')
print(f'prediction accurracy of linear kernel, gamma10, C10 is {np.mean(svm_multihyp2.predict(X_test) == y_test)}')
print(f'prediction accurracy of rbf kernel, gamma10, C10 is {np.mean(svm_multihyp3.predict(X_test) == y_test)}')
```
