```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
feature = wine.get('data')
```

```python
label = wine.get('target')
```

```python
feature_names = wine.get('feature_names')
```

```python
target_names = wine.get("target_names")
```

```python
import pandas as pd
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df
```

```python
df["target"]=label
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
df0=df[df['target']==0]
df1=df[df['target']==1]
df2=df[df['target']==2]
```

```python
plt.figure(figsize=(18,12))
plt.plot(df0['alcohol'])
plt.plot(df1['alcohol'])
plt.plot(df2['alcohol'])
plt.title('Alcohol', fontsize=40)
plt.show()
```

```python
plt.figure(figsize=(6,4))
fig, ax=plt.subplots()
ax.boxplot([df0["alcohol"], df1['alcohol'], df2['alcohol']])
plt.title("Alcohol rates", fontsize=14)
plt.xticks(range(1,4), ["class_0", "class_1", "class_2"])
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df0["alcohol"], alpha=0.5)
plt.hist(df1["alcohol"], alpha=0.5)
plt.hist(df2["alcohol"], alpha=0.5)
plt.title("Alcohol", fontsize=14)
plt.legend(['class_0', 'class_1', 'class_2'], bbox_to_anchor=(1.3,1))

plt.show()
```

```python
plt.figure(figsize=(20,4))
ratio=[len(df0), len(df1), len(df2)]
labels = ['class_0', 'class_1', 'class_2']
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.legend(['class_0', 'class_1', 'class_2'], bbox_to_anchor=(1.3,1))
plt.show()
```
